```python
%load_ext autoreload
%autoreload 2

import sys
import os

ROOT_DIR = os.path.abspath('..')
sys.path.append(ROOT_DIR)
os.chdir(ROOT_DIR)
```

```python
from neuron.neurons import User, Neuron, ChitChat, ChitChatManager
```

```python
llm_config={
    "config_list": [
        {
            "client": "groq",
            "temperature": 0.0,
            "model": "llama-3.3-70b-versatile",
            "api_key": os.getenv("GROQ_API_KEY")
        }
    ]
}
user = User(
    human_input_mode="ALWAYS",
    description="A human user capable of working with Autonomous AI Agents.",

    )

conversational = Neuron(
    name="conversational",
    llm_config=llm_config,
    system_message="You are an expert chitchatter.Call recommender agent If he ask for recommendation, ",
    description =  "This agent acts as a seller"
)

recommendation = Neuron(
    name="recommender",
    llm_config=llm_config,
    system_message="""
    Every time the user ask for a recommonedation you have to list the items below:
     1. Shoes,
     2. t-shirt,.
     3. jeans.
     """,
    description =  "This agent acts as a recommender."
)

chitchat = ChitChat(agents=[user, conversational, recommendation], messages=[], max_round=100, speaker_selection_method="auto", send_introductions = True)
chichat_manager = ChitChatManager(chitchat=chitchat, llm_config=llm_config)
```

```python
user.initiate_chat(
    chichat_manager, message="Oi tudo bem?"
)
```
